           SDLC-Software Development Lift Cycle
           
           SDLC process is used to design, develop, test and maintain in an effective way.This is a step by step process from an idea in mind to existing and working software.

           There are many steps in SDLC like

           **Requirement Analysis**
           In this phase pruduct manage talk to clint and understand his requirements and features.

           **Designing system**
           In this phase TL or architecture understand feature and choose which technology is going to be use. And design database, Ui,and how software is going to perform is decided by him.

           **Development Team**
           This this work on coding like developing frontend, backend etc.
           they develop all complete software.
         **  Q A tester**
         This team find the bug in the software and make software ready to live.

         **DevOps Team**
         This team make the software live.

         **Maintenance Team**
           This team ensure the smooth running of the software. they maintain the software wherever there is an issue with the software during runtime.


           This is the complete lift cycle of software from development to making it well maintained.
